```python
import pandas as pd
import pgmpy
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination

#Load dataset
data = pd.read_csv('data/healthcare/diabetesBayessianRegressed.csv',sep=';')
#data = pd.read_csv('data/healthcare/diabetesMean.csv',sep=';')
_SavePath="data/healthcare/diabetesKnowledgeBinnedBayeReg_NEW_KEYS.csv"
#_SavePath="data/healthcare/diabetesKnowledgeBinnedMean.csv"

#Bayesian network
model = BayesianNetwork([
    ('Pregnancies', 'Outcome'),
    ('Glucose', 'Outcome'),
    ('BloodPressure', 'Outcome'),
    ('Insulin', 'Outcome'),
    ('BMI', 'Outcome'),
    ('DiabetesPedigreeFunction', 'Outcome'),
    ('Age', 'Outcome')
])

ddDataRange = {'Age':{1:'X<56',2:'56<=X<70',3:'X>70'},
               'BMI':{1:'X<18.5',2:'18.5<=X<25',3:'25<=X<30',4:'X>=30'},
               'Pregnancies':{0:'0',1:'1<=X<3',2:'X>=3'},
               'BloodPressure':{1:'X<80',2:'80<=X<85',3:'85<=X<90',4:'X>=90'},
               'Insulin':{1:'X<16',2:'16<=X<166',3:'X=>166'},
               'Glucose':{1:'X<140',2:'140<=X<200'},
               'DiabetesPedigreeFunction':{1:'<0.2',2:'0.2 <= <0.45',3:'0.45 <= <0.8',4:'0.8 <= <1.2',5:'=> 1.2'}
               }

#Podejście zaproponowane przez specjalistę (lekarz pediatra)
# data.loc[data['Age']<56,('Age')] = 1
# data.loc[(data['Age']>=56) & (data['Age']<70),('Age')] = 2
# data.loc[data['Age']>=70,('Age')] = 3

# data.loc[data['BMI']<18.5,('BMI')] = 1
# data.loc[(data['BMI']>=18.5) & (data['BMI']<25),('BMI')] = 2
# data.loc[(data['BMI']>=25) & (data['BMI']<30),('BMI')] = 3
# data.loc[data['BMI']>=30,('BMI')] = 4

# data.loc[(data['Pregnancies']>0) & (data['Pregnancies']<3),('Pregnancies')] = 1
# data.loc[data['Pregnancies']>=3,('Pregnancies')] = 2

# data.loc[data['Glucose']<140,('Glucose')] = 1
# data.loc[(data['Glucose']>=140) & (data['Glucose']<200),('Glucose')] = 2
# data.loc[data['Glucose']>=200,('Glucose')] = 3
 
# data.loc[data['BloodPressure']<80,('BloodPressure')] = 1
# data.loc[(data['BloodPressure']>=80) & (data['BloodPressure']<85),('BloodPressure')] = 2
# data.loc[(data['BloodPressure']>=85) & (data['BloodPressure']<90),('BloodPressure')] = 3
# data.loc[data['BloodPressure']>=90,('BloodPressure')] = 4

# data.loc[data['Insulin']<16,('Insulin')] = 1
# data.loc[(data['Insulin']>=16) & (data['Insulin']<=166),('Insulin')] = 2
# data.loc[data['Insulin']>166,('Insulin')] = 3

# data.loc[data['DiabetesPedigreeFunction']>1.2,('DiabetesPedigreeFunction')] = 5
# data.loc[(data['DiabetesPedigreeFunction']>0.80) & (data['DiabetesPedigreeFunction']<=1.2),('DiabetesPedigreeFunction')] = 4
# data.loc[(data['DiabetesPedigreeFunction']>0.45) & (data['DiabetesPedigreeFunction']<=0.80),('DiabetesPedigreeFunction')] = 3
# data.loc[(data['DiabetesPedigreeFunction']>0.2) & (data['DiabetesPedigreeFunction']<=0.45),('DiabetesPedigreeFunction')] = 2
# data.loc[data['DiabetesPedigreeFunction']<=0.2,('DiabetesPedigreeFunction')] = 1
#data.to_csv(path_or_buf=_SavePath,sep=';')


#Podejście zaproponowane przez specjalistę (lekarz pediatra)
data.loc[data['Age']<56,('Age')] = 28
data.loc[(data['Age']>=56) & (data['Age']<70),('Age')] = 63
data.loc[data['Age']>=70,('Age')] = 75

data.loc[data['BMI']<18.5,('BMI')] = 15
data.loc[(data['BMI']>=18.5) & (data['BMI']<25),('BMI')] = 22
data.loc[(data['BMI']>=25) & (data['BMI']<30),('BMI')] = 27
data.loc[data['BMI']>=30,('BMI')] = 35

data.loc[(data['Pregnancies']>0) & (data['Pregnancies']<3),('Pregnancies')] = 1
data.loc[data['Pregnancies']>=3,('Pregnancies')] = 4

data.loc[data['Glucose']<140,('Glucose')] = 80
data.loc[(data['Glucose']>=140) & (data['Glucose']<200),('Glucose')] = 170
data.loc[data['Glucose']>=200,('Glucose')] = 250
 
data.loc[data['BloodPressure']<80,('BloodPressure')] = 60
data.loc[(data['BloodPressure']>=80) & (data['BloodPressure']<85),('BloodPressure')] = 83
data.loc[(data['BloodPressure']>=85) & (data['BloodPressure']<90),('BloodPressure')] = 88
data.loc[data['BloodPressure']>=90,('BloodPressure')] = 100

data.loc[data['Insulin']<16,('Insulin')] = 10
data.loc[(data['Insulin']>=16) & (data['Insulin']<=166),('Insulin')] = 90
data.loc[data['Insulin']>166,('Insulin')] = 200

data.loc[data['DiabetesPedigreeFunction']>1.2,('DiabetesPedigreeFunction')] = 2
data.loc[(data['DiabetesPedigreeFunction']>0.80) & (data['DiabetesPedigreeFunction']<=1.2),('DiabetesPedigreeFunction')] = 1
data.loc[(data['DiabetesPedigreeFunction']>0.45) & (data['DiabetesPedigreeFunction']<=0.80),('DiabetesPedigreeFunction')] = 0.6
data.loc[(data['DiabetesPedigreeFunction']>0.2) & (data['DiabetesPedigreeFunction']<=0.45),('DiabetesPedigreeFunction')] = 0.3
data.loc[data['DiabetesPedigreeFunction']<=0.2,('DiabetesPedigreeFunction')] = 0.1
data.to_csv(path_or_buf=_SavePath,sep=';')


#bmi do 25, 25-30, +30
#ciąż 0, 1-2, +3 (lub 0, 1, +2)
#glukoza poniżej 140, 140-199, +200
#ciśnienie krwi poniżej 80, 80-90, powyżej 90 (lub poniżej 80, 80-85, 85-90, powyżej 90)
#insulina poniżej 16, 16-166, powyżej 166
#DiabPediFunc - dobrane na podstawie odwzorowania kształtów ze względu na niedobór informacji
#umożliwiających podziału z punktu widzenia medycyny

#https://www.analyticsvidhya.com/blog/2021/07/diabetes-prediction-with-pycaret/

#Model fit
model.fit(data, estimator=MaximumLikelihoodEstimator)

#Inference
infer = VariableElimination(model)

#Features for prediction
evidence = {'Age':1,'BMI': 2, 'Pregnancies': 0, 'Glucose': 1, 'BloodPressure': 1, 'Insulin': 1}

#Prediction based on features set
diabetes_prob = infer.query(['Outcome'], evidence=evidence)
for x in evidence:
    print(x,":",ddDataRange[x][evidence[x]])
print(diabetes_prob)
```
